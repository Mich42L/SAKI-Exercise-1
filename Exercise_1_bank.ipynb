```python
from sklearn.model_selection import train_test_split
# get_dummies: Converts categorical variable into dummy/indicator variables.
# not that important for the task!

X_train, X_test, y_train, y_test = train_test_split(pd.get_dummies(bank[['Buchungstext', 'Kontonummer']]),pd.DataFrame(bank['label']),test_size=0.3,random_state=None)
#random state choosen, will be determined later

print(X_train.info())
print(X_train.head())
```

**1 - Create a Gaussian Classifier**
***
**2- Train the model using the training sets**
***
**3- Predict Output!**

```python
from sklearn import metrics
# Create a Gaussian Classifier
model = GaussianNB()

#determine the best random state
i=0
best={"random_state":0,"accuracy":0}
all_scores = np.array([])
while (i<101):
    #
    X_train, X_test, y_train, y_test = train_test_split(pd.get_dummies(bank[['Buchungstext', 'Kontonummer']]),pd.DataFrame(bank['label']),test_size=0.3,random_state=i)

# Train the model using the training sets 
    model.fit(X_train, y_train.values.ravel())
#Predict Output 
    y_pred= model.predict(X_test)
# Model Accuracy
    score=metrics.accuracy_score(y_test, y_pred)
    
    if(score>best["accuracy"]):
        best["random_state"]=i
        best["accuracy"]=score
    all_scores= np.append(all_scores,score)   
    i=i+1
    
avg = np.average(all_scores)
#print(all_scores)
#print("Average from 0 to 100: ")
#average was useless, to show the value, just delete the comment
#print(avg)
print(best)
```
